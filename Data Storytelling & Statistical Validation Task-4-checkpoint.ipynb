## Hypothesis Testing

H0: There is no significant difference in average sales between selected categories.

H1: There is a significant difference in average sales between selected categories.

```python
#  Check category counts
print(df['Category'].value_counts())
```

```python
# Select top 2 categories with highest data
top_categories = df['Category'].value_counts().index[:2]
print("Selected Categories:", top_categories)
```

```python
#  Create groups
cat1 = df[df['Category'] == top_categories[0]]['Total_Sales'].dropna()
cat2 = df[df['Category'] == top_categories[1]]['Total_Sales'].dropna()
```

```python
#  Check sizes
print("Sample sizes:", len(cat1), len(cat2))
```

```python
#  Perform T-test
t_stat, p_value = ttest_ind(cat1, cat2, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
#
alpha = 0.05

if p_value < alpha:
    print("Reject H0 → Significant difference in sales")
else:
    print("Fail to reject H0 → No significant difference")
```

## Final Result Interpretation

The computed p-value (0.7937) is greater than the significance level (0.05).

Therefore, we fail to reject the null hypothesis (H0).

This indicates that there is no statistically significant difference in average sales between the selected categories.

Hence, any observed difference in sales is likely due to random variation rather than a meaningful business effect.

## Business Conclusion

The analysis shows that category-wise differences in sales are not statistically significant.

This suggests that focusing only on product categories may not be sufficient to drive higher sales.

Businesses should explore other influencing factors such as customer behavior, pricing strategies, and marketing efforts to improve overall performance.
